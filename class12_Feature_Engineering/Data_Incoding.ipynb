```python
import pandas as pd 
import random
```

```python
colors =  ["red" ,"green", "blue" , "black"]
df = pd.DataFrame({
    'color' : [random.choice(colors) for _ in range(50)]
})
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['color']]).toarray()
```
